```python
import random
import re
from collections import defaultdict
```

```python
text = """
Artificial Intelligence is transforming the world.
Machine learning allows computers to learn from data.
Deep learning is a powerful subset of machine learning.
Artificial intelligence is used in healthcare, finance, and robotics.
Data is the fuel of modern AI systems.
"""
```

```python
def preprocess(text):
    
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    words = text.split()
    
    return words

words = preprocess(text)

print(words[:20])
```

```python
def build_markov_chain(words):

    markov_chain = defaultdict(list)

    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]

        markov_chain[current_word].append(next_word)

    return markov_chain


markov_model = build_markov_chain(words)

print(dict(list(markov_model.items())[:5]))
```

```python
def generate_text(model, start_word, length=20):

    current_word = start_word
    sentence = [current_word]

    for _ in range(length):

        next_words = model.get(current_word)

        if not next_words:
            break

        next_word = random.choice(next_words)

        sentence.append(next_word)
        current_word = next_word

    return " ".join(sentence)
```

```python
generated = generate_text(markov_model, start_word="artificial", length=20)

print("Generated Text:\n")
print(generated)
```

```python
for i in range(5):
    print(generate_text(markov_model, "machine", 15))
    print()
```

```python
import pandas as pd

data = []

for word, next_words in markov_model.items():
    for next_word in next_words:
        data.append((word, next_word))

df = pd.DataFrame(data, columns=["Current Word", "Next Word"])

df.head()
```
